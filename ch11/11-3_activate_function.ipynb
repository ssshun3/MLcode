# 活性化関数
- 様々な活性化関数を実装し、そのグラフを表示してみよう

## 1. ライブラリの読み込み

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. ステップ関数
$
h(x) = \left\{
\begin{array}{ll}
1 & x\geq0 \\
0 & x< 0
\end{array}
\right.
$
- 入力が閾値以上であれば1を出力、そうでなければ0を出力

```python
# 関数の定義
def step(x):
    return np.array(x>0).astype(np.int32)
    
# 横軸の値を生成
x = np.arange(-5,5,0.1)
# 縦軸の値（関数の出力）を計算
y = step(x) 

# グラフの表示
sns.lineplot(x=x,y=y)
# ラベルの設定
plt.xlabel("x")
plt.ylabel("y")
# 表示範囲の調整
plt.ylim([-0.1,1.1])
plt.show()
```

## 3. シグモイド関数
### $y = \frac{1}{1+\mathrm{exp}(-x)}$
- 任意の実数を入力として、0から1の間の値を出力
- 導関数を簡単に計算することができる
    - $\frac{\partial y}{\partial x}= y(1-y)$
    - 逆伝播計算の際に有用

```python
# 関数の定義
def sigmoid(x):
    return 1 / (1+ np.exp(-x))

# 縦軸の値（関数の出力）を計算
y = sigmoid(x) 

# グラフの表示
sns.lineplot(x=x,y=y)
# ラベルの設定
plt.xlabel("x")
plt.ylabel("y")
# 表示範囲の調整
plt.ylim([-0.1,1.1])
plt.show()
```

## 4. tanh関数
### $y = \frac{e^x-e^{-x}}{e^x+e^{-x}}$
- シグモイド関数と同様にS字形のカーブをもつ
- 出力範囲が-1から1の範囲になるため、0を中心とした出力が得られる
    - ニューラルネットワークの中間層に用いると、学習にあたって都合が良い

```python
# 関数の定義
def tanh(x):
    return np.tanh(x)

# 縦軸の値（関数の出力）を計算
y = tanh(x) 

# グラフの表示
sns.lineplot(x=x,y=y)
# ラベルの設定
plt.xlabel("x")
plt.ylabel("y")
# 表示範囲の調整
plt.ylim([-2,2])
plt.show()
```

## 5. [演習] ReLU関数
$\mathrm{ReLU(x)} = \max{(x, 0)}$
- 入力$x$が0以上ならば$x$を出力し、0未満ならば0を出力する
- 勾配消失の対策として有用

```python
# 関数の定義
def relu(x):
    x = np.maximum(0, x)
    return x

# 縦軸の値（関数の出力）を計算
y = relu(x) 

# グラフの表示
sns.lineplot(x=x,y=y)
# ラベルの設定
plt.xlabel("x")
plt.ylabel("y")
# 表示範囲の調整
plt.ylim([-5.0,5])
plt.show()
```
