## 2. データの読み込み
- 目的変数
    - funded_amount：合計出資額（Kiva から現地エージェントへの送金額）[USドル]
        - 貸し手から Kiva に集まった資金の合計金額
        - 回帰問題のみで用いる（**分類問題の場合は削除**）
    - **funded_state：融資予定額と同額の出資金が期限内に集まったか（funded$\geq$loanならば 1)**
        - 集まらなかった場合、出資された資金は貸し手に返金される
        - **分類問題のみで用いる**（回帰問題の場合は削除）
- 説明変数
    - loan_amount：融資予定額（現地エージェントが借り手に支払った金額）[USドル]
    - activity：具体的なカテゴリ
    - sector：抽象的なカテゴリ
    - use：借入目的
    - country_code：国を表す記号
    - country：国
    - region：地域
    - currency：通貨
    - partner_id：パートナー団体のID
    - posted_time：現地エージェントがKivaに融資情報を掲載した日時
    - disbursed_time：現地エージェントが借り手への融資を実行した日時
    - funded_time：出資者（貸し手）が融資を全額完了した日時
    - term_in_months：融資を受けた期間 [ヶ月]
    - lender_count：出資者の合計人数 [人]
    - tag：借入目的に関するタグ
    - borrower_genders：借り手の性別（1文字が1人を表す）
    - repayment_interval：返済の間隔（不定期、月極、週次、一括）

## 4. データの前処理
前処理については DAY2 で詳しく扱うため、ここでは最低限の操作のみを行う
- 不要な列の削除
- カテゴリ変数のダミー変数化

### 4-1. 不要な列の削除
- 欠損値を含む列を全て削除
    - use, region, partner_id, disbursed_time, funded_time, tags, borrower_genders
    - 欠損値処理のひとつ（詳しくは DAY2 で扱う）
    
- 目的変数そのものの情報を含んでいる変数を削除
    - funded_amount は、**funded_state** を作成する際に使っているため削除
    - posted_time, lender_count もリーケージ（※）の恐れがあるため削除
- 重複した情報をもつ変数を削除
    - activity は sector をさらに細分化したものであるため、sector を残して activity を削除
    - country, country_code, currency は全て地域を表す変数であるため、currency 以外を削除
    
※リーケージ…本来得られるはずのないデータをモデルの学習時に使用してしまうこと


```python
removed_cols = ['use', 'region', 'partner_id', 'disbursed_time', 'funded_time', 'tags', 'borrower_genders',\
  'funded_amount', 'posted_time', 'lender_count', 'activity', 'country', 'country_code']
ndf = df.drop(columns=removed_cols)
ndf.head()
```

```python
# 欠損値がないことを確認
ndf.info()
```

### 4-2. カテゴリ変数のダミー変数化
詳しい使い方は 2_notebook/appendix/dummy_variable.ipynb を参照
- 数値型の変数：loan_amount, term_in_months, **funded_state**
- カテゴリ変数：sector, currency, repayment_interval

```python
# ダミー変数化
ndf_ = pd.get_dummies(ndf, drop_first=True)

# 列名の確認
ndf_.columns
```

## 5. ロジスティック回帰

### 5-1. モデルの構築・学習
- scikit-learn の [SGDClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDClassifier.html) を使うことで簡単にロジスティック回帰を実装できる

[参考] [LogisticResression](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) でもモデルを実装可能だが、最適化手法の設定が必要、多クラス分類を行えないなどの違いがある

```python
# 説明変数
X = ndf_.drop(columns=['funded_state'])

# 目的変数
y = ndf_['funded_state']

# ロジスティック回帰による分類器のインスタンスを作成
lr_clf = SGDClassifier(
    loss='log_loss', max_iter=10000, fit_intercept=True, 
    random_state=1234, tol=1e-3) 

# 学習を実行
lr_clf.fit(X, y) 
```

### 5-2. モデルの評価
- ここでは、あくまでも学習時に使ったデータに対する評価値（訓練性能）を求めていることに注意
- 学習に使ったデータと同じデータを評価に使うと、汎化性能（学習に使った以外のデータにも適応できるか？）を評価できない
    - そのため、通常は学習用と評価用のデータは分けておく必要がある
    - この点については DAY2 で詳しく扱う


```python
# 予測値の算出
y_predicted = lr_clf.predict(X)

# 評価指標の値を算出
acc = accuracy_score(y, y_predicted)
precision = precision_score(y, y_predicted)
recall = recall_score(y, y_predicted)
f_1 = f1_score(y, y_predicted)

print(f'正解率: {acc:.3}')
print(f'Precision: {precision:.3}')
print(f'Recall: {recall:.3}')
print(f'F1: {f_1:.3}')
```
